## Показать понедельную динамику посещаемости сайта за сентябрь-ноябрь 2016. Показатели: визиты, просмотры, среднедневное кол-во визитов на посетителя

```python
tmp = data[(data['Интервал дат визита'] >= datetime.datetime(2016, 9, 1)) &
         (data['Интервал дат визита']<= datetime.datetime(2016, 11, 30))]
tmp = tmp.loc[:, ['Визиты', 'Посетители']].groupby(tmp['Интервал дат визита'].dt.isocalendar().week).agg(['sum', 'mean'])
tmp.columns = ["_".join(pair) for pair in tmp.columns]
tmp = tmp.reset_index()
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.set_title('Динамика Визитов и Посетителей по неделям')
_ = sns.lineplot(data = tmp.loc[:,['week', 'Визиты_sum', 'Посетители_sum']].melt('week', var_name='activity', value_name='value'),
                 x='week',
                 y='value',
                 hue='activity',
                 ax=ax)
```

```python
tmp['Среднедневное'] = tmp['Визиты_mean']/tmp['Посетители_mean']
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.set_title('Динамика среднедневного по неделям')
_ = sns.lineplot(data = tmp, x='week', y='Среднедневное',ax=ax)
```

## Рассчитать по каждой отчётной неделе (пн-вс) изменение за неделю и за год (неделя 2016 г. к той же неделе 2015 г.)

```python
tmp = data[(data['Интервал дат визита'] >= datetime.datetime(2015, 1, 1)) &
         (data['Интервал дат визита']<= datetime.datetime(2015, 12, 31))]
tmp = tmp.loc[:, ['Визиты', 'Посетители']].groupby(tmp['Интервал дат визита'].dt.isocalendar().week).agg(['sum', 'mean'])
tmp.columns = ["_".join(pair) for pair in tmp.columns]
tmp['Среднедневное'] = tmp['Визиты_mean']/tmp['Посетители_mean']
tmp.drop(['Визиты_mean', 'Посетители_mean'], inplace=True, axis=1)
```

```python
tmp1 = data[(data['Интервал дат визита'] >= datetime.datetime(2016, 1, 1)) &
         (data['Интервал дат визита']<= datetime.datetime(2016, 12, 31))]
tmp1 = tmp1.loc[:, ['Визиты', 'Посетители']].groupby(tmp1['Интервал дат визита'].dt.isocalendar().week).agg(['sum', 'mean'])
tmp1.columns = ["_".join(pair) for pair in tmp1.columns]
tmp1['Среднедневное'] = tmp1['Визиты_mean']/tmp1['Посетители_mean']
tmp1.drop(['Визиты_mean', 'Посетители_mean'], inplace=True, axis=1)
```

```python
res = tmp1 - tmp
res = res.reset_index()
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.set_title('Относительная динамика 2015 и 2016, по Визитам и Посетителям')
plot = sns.lineplot(data = res.loc[:,['week', 'Визиты_sum', 'Посетители_sum']].melt('week', var_name='activity', value_name='value'),
                 x='week',
                 y='value',
                 hue='activity',
                 ax=ax)
_=plot.axhline(y=0, color='red')
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.set_title('Относительная динамика между 2015 и 2016, среднедневного по неделям')

plot = sns.lineplot(data = res, x='week', y='Среднедневное',ax=ax)
_=plot.axhline(y=0, color='red')
```

## Построить помесячный прогноз трафика проекта на 2017 год (по визитам)

Насколько я понял имеется ввиду дефолтный для Excel прогноз по алгоритму экспоненциального сглаживания ETS(A, A, A). Я не знаком с этим алгоритмом, сделал по тому, что смог понять за пару часов.

```python
from statsmodels.tsa.exponential_smoothing.ets import ETSModel
```

```python
month_data = data
month_data.index = month_data['Интервал дат визита']
month_data = month_data.loc[:, ['Визиты',]].groupby(pd.Grouper(freq='M')).sum()
month_data = month_data.squeeze()
```

```python
model = ETSModel(
    month_data,
    error="add",
    trend="add",
    seasonal="mul",
    damped_trend=True,
    seasonal_periods=12
)
fit = model.fit()
```

```python
fig, ax = plt.subplots(figsize=(16,9))

month_data.plot(label="data")
fit.fittedvalues.plot(label="statsmodels fit")

plt.ylabel("Кол-во Визитов")
_=plt.legend()
```

```python
pred = fit.get_prediction(start="2017.01.01", end="2017.12.31")
predicted_df = pred.summary_frame(alpha=0.05)
```

```python
predicted_df.columns = [0, 'mean_numerical', 'pi_lower', 'pi_upper']
month_data_with_predict = pd.concat([month_data, predicted_df])
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.set_title('Динамика визитов по месяцам, с предсказанием на 2017г')
_=month_data_with_predict.plot(ax=ax)
```
